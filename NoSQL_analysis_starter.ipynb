### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Find the latitude and longitude of "Penang Flavours"
query = {'BusinessName': 'Penang Flavours'}

result = establishments.find_one(query)
# Extract latitude and longitude
if "geocode" in result:
    longitude = result["geocode"]["longitude"]
    latitude = result["geocode"]["latitude"]

   # print the retrieved data
    print(f"Longitude for Penang Flavours is: {longitude}")
    print(f"Latitude for Penang Flavours is : {latitude}")
else:
    print("geocode field not found")
        
```

```python
print(longitude)
print(latitude)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude =  51.490142
longitude = 0.08384

query = {
    'RatingValue': {'$eq': 5},
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}
}
#Picking only the most important fields to be displayed in the final dataframe
fields = {'geocode': 1, 'BusinessName': 1, 'AddressLine1': 1,'scores.Hygiene': 1, 'RatingValue': 1, 'BusinessType':1}
sort = [('scores.Hygiene', 1)]
limit = 5

# Cast the results as a list and save them to a variable
result = list(establishments.find(query, fields).sort(sort).limit(limit))
         
# Use count_documents to display the number of documents in the result
count3 = establishments.count_documents(query)
print(f"There are {count3} establishments nearest to the new restaurant added, 'Penang Flavours' with a RatingValue of 5, sorted by lowest hygiene score.")


# Pretty print the results
print("Five such establishments are as following: ")
pprint(result)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Checking the data type of Hygiene within scores dictionery
# Find a sample document
sample_document3 = establishments.find_one()

# Check if the "scores" field exists in the document
if "scores" in sample_document3:
    # Access data inside the "scores" dictionary and Hygiene field
    Hygiene_data_type = type(sample_document3["scores"]["Hygiene"])
  
    # Print the retrieved data
    print(f"Hygiene's datattype is: {Hygiene_data_type}")
   
else:
    print("'Scores' and 'Hygiene' fields are not found.")
```

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq': 0 }}}

# 2. Groups the matches by Local Authority
group_query = { '$group': {'_id': "$LocalAuthorityName", 'Count': {'$sum':1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'Count':-1}}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

```

```python
# Run the pipeline through the aggregate method and save the results to a variable
results = list(establishments.aggregate(pipeline))
```

```python
# Print the number of establishments by Local Authority Name in the result
print("Number of establishments by Local Authority Name in the result: ", len(results))
print("The first ten results are: ")
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))

# Display the first 10 rows of the DataFrame
result_df.head(10)
```
